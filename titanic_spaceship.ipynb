DATA PREPROCESSING

```python
data[['cabin1','cabin2','cabin3']]=data['Cabin'].str.split('/',expand=True)
data=data.drop(['Cabin'],axis=1)
data
```

```python
X=data
X
y
```

```python
X.info()
```

```python
X.isnull().mean()*100
```

```python
num_cols=X.select_dtypes(include=['float64']).columns.tolist()
object_cols=X.select_dtypes(include='object').columns.tolist()
num_cols
object_cols.pop(0)
object_cols.pop(4)
object_cols.pop(4)
object_cols
```

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(strategy='mean')
X[num_cols]=imputer.fit_transform(X[num_cols])
X[num_cols].isna().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
scaler=StandardScaler()
X[num_cols]=scaler.fit_transform(X[num_cols])
```

```python
encoder=OneHotEncoder(sparse=False,handle_unknown='ignore')
encoder.fit_transform(X[object_cols])
encoder.categories_
```

```python
encoded_col=list(encoder.get_feature_names(object_cols))
encoded_col
```

```python
X[encoded_col]= encoder.transform(X[object_cols])
X[encoded_col]
```

```python
X=X[encoded_col+num_cols]
X
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
test=pd.read_csv('test.csv')
test
```

```python
test['cabin3']= test.apply(lambda x : str(x['Cabin'])[-1:], axis=1)
test[num_cols]=imputer.fit_transform(test[num_cols])
test[encoded_col]=encoder.transform(test[object_cols])
```

```python
X_test=test[encoded_col+num_cols]
X_test
```

```python
#generating csv for submission
predict=model.predict(X_test)
predict
```

```python
submisson_csv=pd.DataFrame(columns=['PassengerId','Transported'])
submisson_csv['PassengerId']=test['PassengerId']
submisson_csv['Transported']=predict
submisson_csv
```
